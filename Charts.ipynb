```python
# 散点图
######---------------------------
def Scatter_Plot(x_train, y_train, title, xAxisName, yAxisName,ax):
    #图表标题
    plt.title(title)

    #绘制
    plt.plot(x_train, y_train,label='curve1',color='#706d6a', linewidth=2, linestyle='--', alpha=0.3) #画曲线
    # plt.plot(x_train, y_train, '.') #画散点图方式1
    plt.scatter(x_train, y_train, s=50,facecolor='#706d6a') #画散点图方式2 
    ax.legend(); #曲线名称显示

    #坐标轴名称设置
    ax.set_xlabel('$'+xAxisName+'$')
    ax.set_ylabel('$'+yAxisName+'$')

    #是否开启网格
    ax.grid(True,alpha=0.3);

    #显示图表
    plt.show()
```

```python
# 有预测曲线的散点图
######---------------------------
def Scatter_Regressed(x_train, y_train, y_result, title, xAxisName, yAxisName,ax):
    #图表标题
    plt.title(title)

    #绘制
    plt.plot(x_train, y_result,label='curve1',color='#e47364', linewidth=2) #画曲线
    # plt.plot(x_train, y_train, '.') #画散点图方式1
    plt.scatter(x_train, y_train, s=50,facecolor='#706d6a') #画散点图方式2 
    ax.legend(); #曲线名称显示

    #坐标轴名称设置
    ax.set_xlabel('$'+xAxisName+'$')
    ax.set_ylabel('$'+yAxisName+'$')

    #是否开启网格
    ax.grid(True,alpha=0.3);

    #显示图表
    plt.show()
```

```python
# 简单条形图
######-------------------
def bar():
    fig, ax = plt.subplots()

    #图表标题
    plt.title('house price')

    #坐标轴名称设置
    ax.set_ylabel('$price$')

    #绘制
    bar_labels = ['red', 'blue', 'red', 'orange']
    bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']
    x = ['apple', 'blueberry', 'cherry', 'orange']
    y = [5, 20, 15, 25]
    ax.bar(x, y, label=bar_labels, color=bar_colors,alpha=0.5)
    ax.legend(title='Fruit color')

    #显示图表
    plt.show()
```

```python
def pie():
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    #使得X/Y轴的间距相等
    ax.axis('equal')
    #准备数据
    x = ['C', 'C++', 'Java', 'Python', 'PHP']
    percentage = [23,17,35,29,12]
    #绘制饼状图
    ax.pie(percentage, labels = x,autopct='%1.2f%%')
    ax.legend(title='languages')

    #显示图表
    plt.show()
```

```python
# 0，1分类图
def Binary_Classification(X, y, ax, pos_label="y=1", neg_label="y=0", s=80, loc='best' ):
    pos = y == 1
    neg = y == 0
    pos = pos.reshape(-1,)  #work with 1D or 1D y vectors
    neg = neg.reshape(-1,)

    # Plot examples
    ax.scatter(X[pos, 0], X[pos, 1], marker='x', s=s, c = 'red', label=pos_label)
    ax.scatter(X[neg, 0], X[neg, 1], marker='o', s=s, label=neg_label, facecolors='none', edgecolors='blue', lw=3)
    ax.legend(loc=loc)

    ax.figure.canvas.toolbar_visible = False
    ax.figure.canvas.header_visible = False
    ax.figure.canvas.footer_visible = False
    plt.show()
```
